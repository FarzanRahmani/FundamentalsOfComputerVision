```python
import numpy as np
import cv2
import glob
import matplotlib.pylab as plt
import os
```

<h3><font color = "#079e05"> Read input images from 1 to 8 and show them in a row together </font></h3>

```python
image_paths = glob.glob('images/Q2/*.jpg') # https://docs.python.org/3/library/glob.html 

# initialized a list of images
imgs = []
for i in range(len(image_paths)):
    imgs.append(cv2.imread(image_paths[i]))
    # imgs[i] = cv2.resize(imgs[i], (0,0), fx=0.5, fy=0.5)
    # this is optional if your input images isn't too large
    # you don't need to scale down the image
    # in my case the input images are of dimensions 3000x1200
    # and due to this the resultant image won't fit the screen
    # scaling down the images 

print('Number of images: ', len(imgs))
print('Image1 shape: ', imgs[0].shape)
print('Image2 shape: ', imgs[1].shape)
print('Image3 shape: ', imgs[2].shape)
print('Image4 shape: ', imgs[3].shape)
print('Image5 shape: ', imgs[4].shape)
print('Image6 shape: ', imgs[5].shape)
print('Image7 shape: ', imgs[6].shape)
print('Image8 shape: ', imgs[7].shape)

# cv2.imshow('1',imgs[0])
# cv2.imshow('2',imgs[1])
# cv2.imshow('3',imgs[2])
# cv2.imshow('4',imgs[3])
# cv2.imshow('5',imgs[4])
# cv2.imshow('6',imgs[5])
# cv2.imshow('7',imgs[6])
# cv2.imshow('8',imgs[7])
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

<h3><font color = "#079e05"> Initialize OpenCV's image sticher object and then perform the image stitching on input images </font></h3>

```python
# naive approach
# output = np.concatenate((imgs[0],imgs[1],imgs[2],imgs[3],imgs[4],imgs[5],imgs[6],imgs[7]), axis=1)
# cv2.imwrite('output.jpg', output)

stichy = cv2.Stitcher.create()
status, result = stichy.stitch(imgs)
if status == cv2.STITCHER_OK:
    # checking if the stitching procedure is successful
    # .stitch() function returns a true value if stitching is 
    # done successfully
    print("Stitching completed successfully")
    cv2.imwrite('images/Q2/output/output.jpg', result)
else:
    print("Error during stitching")
```

<h3><font color = "#079e05"> Show the result </font></h3>

```python
cv2.imshow('result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()

# plot the result with matplotlib
plt.figure(figsize=(20, 20))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```
